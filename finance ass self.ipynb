```python
import pandas as pd
import yfinance as yf
from matplotlib import pyplot as plt
```

```python
data = yf.download("ZOMATO.NS", start = '2023-01-01', end = '2023-04-01')
```

```python
data.head()
```

```python
print(data.head())
```

```python
data['Simple Moving Average'] = data['Close'].rolling(10).mean()
```

```python
data
```

```python
data.shape
```

```python
data['Exponential Moving Average'] = data['Close'].ewm(span=10, adjust=True).mean()
```

```python
data
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data["Close"], label="Close")
plt.plot(data["Simple Moving Average"], label="SMA")
plt.plot(data["Exponential Moving Average"], label="EMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Moving Average Indicators")
plt.legend()
plt.grid(True)
plt.show()
```
